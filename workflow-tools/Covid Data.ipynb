```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
hospital_data = pd.read_csv('../model/hospital_edited_locs.csv')
hospital_data = hospital_data.drop("V1",axis=1)
hospital_data
```

```python
WA = hospital_data.loc[hospital_data["location_name"]=="Washington",:]
WA
```

```python
NY = hospital_data.loc[hospital_data["location_name"]=="New York",:]
NY
```

```python
MA = hospital_data.loc[hospital_data["location_name"]=="Massachusetts",:]
MA
```

```python
IL = hospital_data.loc[hospital_data["location_name"]=="Illinois",:]
IL
```

```python
CA = hospital_data.loc[hospital_data["location_name"]=="California",:]
CA
```

```python
USA = hospital_data.loc[hospital_data["location_name"]=="United States of America",:]
USA
```

```python
plt.plot(USA["date"],USA['admis_mean'],label="USA")
plt.plot(NY["date"],NY['admis_mean'],label="New York")
plt.plot(IL["date"],IL['admis_mean'],label="Illinois")
plt.plot(WA["date"],WA['admis_mean'],label="Washington")
plt.plot(CA["date"],CA['admis_mean'],label="California")
plt.plot(MA["date"],MA['admis_mean'],label="Massachusetts")
plt.xlim(60,105)
plt.ylim(0,11000)
plt.xticks(np.arange(60,105,5))
plt.xticks(rotation=45)
plt.title("New Covid-19 Cases in USA")
plt.xlabel("Date")
plt.ylabel("# of Cases")
plt.legend()
plt.grid()
plt.savefig("../view/static/img/COVID19-CASES-USA.png")
plt.show()
```

```python
plt.plot(NY["date"],NY['admis_mean'],label="# of New Cases")
plt.plot(NY["date"],NY['deaths_mean'],label="# of Deaths")
plt.plot(NY["date"],NY['totdea_mean'],label="# of Total Deaths")
plt.xlim(60,105)
plt.ylim(0,13000)
plt.xticks(np.arange(60,105,5))
plt.xticks(rotation=45)
plt.title("Covid-19 Cases in NY State")
plt.xlabel("Date")
plt.ylabel("# of People")
plt.legend()
plt.grid()
plt.savefig("../view/static/img/COVID19-CASES-NY.png")
plt.show()
```
