```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
from datetime import date
```

```python
# import first csv
indeed_jobs = pd.read_csv('../model/indeed_cleaned.csv')

indeed_jobs
```

```python
# make posted date negative
indeed_jobs['posted'] = -indeed_jobs['posted']

indeed_jobs
```

```python
# group by posted date
groupby_posted = indeed_jobs.groupby(['posted']).count()
groupby_posted
```

```python
# create graph of jobs by days ago posted
plt.plot(groupby_posted['job_title'])
plt.xlim(-30,0)
plt.ylim(0,7250)
plt.xticks(np.arange(-30,1,2))
plt.yticks(np.arange(0,7250,500))
plt.title(f"Number of Job Postings as of {indeed_jobs['date'][0]}")
plt.xlabel("Days Ago")
plt.ylabel("# of Postings")
plt.grid()
plt.savefig("../view/static/img/JOB-POSTINGS-USA.png")
plt.show()
```

### The spike at 17 days ago comes from California

```python
# import second csv
covid_data = pd.read_csv('../model/hospital_edited_locs.csv')

covid_data
```

```python
# convert date column to datetime
covid_data['date'] = pd.to_datetime(covid_data['date'])
indeed_jobs['date'] = pd.to_datetime(indeed_jobs['date'])
```

```python
# create new column of days ago posted 
days_ago = []

for date in covid_data['date']:
    days_ago_dt = -(indeed_jobs['date'][0] - date)
    days_ago.append(days_ago_dt.days)

print(days_ago)
```

```python
# add new column to second csv
covid_data['days ago'] = days_ago

covid_data
```

```python
# sort second csv by location
USA = covid_data.loc[covid_data["location_name"]=="United States of America",:]
USA
```

```python
# create graph of jobs by days ago posted along with covid cases
plt.plot(groupby_posted['job_title'], label = 'Job Postings')
plt.plot(USA["days ago"],USA['admis_mean'],label="USA New Covid Cases")
plt.xlim(-30,0)
plt.ylim(0,10700)
plt.xticks(np.arange(-30,1,2))
plt.yticks(np.arange(0,10700,500))
plt.title(f"Number of Job Postings vs New Covid Cases as of {indeed_jobs['date'][0].strftime('%m/%d')}")
plt.xlabel("Days Ago")
plt.ylabel("# of Postings")
plt.legend()
plt.grid()
plt.savefig("../view/static/img/JOBS-COVID-USA.png")          
plt.show()
```

### The spike at 17 days ago comes from California

```python
# drop all NAs to sort by location
indeed_drop = indeed_jobs.drop(['Unnamed: 0','salary'],axis=1)
indeed_dropna = indeed_drop.dropna()
indeed_dropna
```

```python
indeed_dropna[['city', 'state']] = indeed_dropna["location"].apply(lambda x: pd.Series(str(x).split(", ")))
indeed_dropna
```

```python
indeed_dropna[['state_abb', '1', '2', '3', '4']] = indeed_dropna["state"].apply(lambda x: pd.Series(str(x).split(" ")))
indeed_dropna
```

```python
indeed_dropped = indeed_dropna.drop(['location','state','1','2','3','4'], axis=1)
indeed_dropped
```

```python
# check available states
jobs_by_state = indeed_dropped.groupby(['state_abb']).count()
jobs_by_state
```

```python
NY_jobs = indeed_dropped.loc[indeed_dropped["state_abb"]=="NY",:]
NY_jobs_grouped = NY_jobs.groupby(['posted']).count()
NY_jobs_grouped
```

### New York's first cases of Covid started 37 days ago.

```python
NY_covid = covid_data.loc[covid_data["location_name"]=="New York",:]
NY_covid[70:]
```

```python
# create graph of jobs by days ago posted along with covid cases
plt.plot(NY_jobs_grouped['job_title'], label = 'NY Job Postings')
plt.plot(NY_covid["days ago"],NY_covid['admis_mean'],label="NY New Covid Cases")
plt.xlim(-30,0)
plt.ylim(0,2600)
plt.xticks(np.arange(-30,1,2))
plt.yticks(np.arange(0,2600,200))
plt.title(f"Number of NY Job Postings vs New Covid Cases as of {indeed_jobs['date'][0].strftime('%m/%d')}")
plt.xlabel("Days Ago")
plt.ylabel("# of Postings")
plt.legend()
plt.grid()
plt.savefig("../view/static/img/JOBS-COVID-NY.png")          
plt.show()
```

## Warning: Even though we placed job postings and Covid cases on top of each other we should consider prior Covid spikes and dips as affecting future job postings. So, the dip in Covid cases 19 days ago in CA might account for the spike in job postings 17 days, and the subsequent spike in Covid cases 17 days ago might account for the dip in job postings 13 days ago.
